```python
import pandas as pd
df = pd.read_csv("Inflation_data.csv", skiprows=1)
df.head()
```

```python
df.columns
```

```python
df.shape
```

```python
df = df[["Year", "Month", "Combined"]]
df.head()
```

```python
df.dtypes
```

```python
yearly_cpi = df.groupby("Year")["Combined"].mean()
yearly_cpi
```

```python
inflation = yearly_cpi.pct_change() * 100
inflation
```

```python
import matplotlib.pyplot as plt
```

```python
inflation.plot()
plt.title("India Inflation Trend (2015–2025)")
plt.ylabel("Inflation Rate (%)")
plt.show()
```
